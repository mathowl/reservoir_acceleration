```python
import numpy as np
import shelve
import matplotlib.pyplot as plt

```

```python
for N in [10,100,1000,2500,5000,10_000]:
    
    with shelve.open('shelve/sol_gpu.shelve', 'r') as shelf:
        Yg = shelf[f'{N}']
    with shelve.open('shelve/sol_numba-vanilla.shelve', 'r') as shelf:
        Ynv = shelf[f'{N}']
    with shelve.open('shelve/sol_numba-parallel.shelve', 'r') as shelf:
        Ynp = shelf[f'{N}']
    with shelve.open('Base/sol_base.shelve', 'r') as shelf:
        Yb = shelf[f'{N}']
    
    
    plt.title(f'Error with respect to base N={N}')
    plt.plot(np.linalg.norm(Ynv-Yb,axis=1))
    plt.plot(np.linalg.norm(Ynp-Yb,axis=1))
    plt.plot(np.linalg.norm(Yg-Yb,axis=1))
    plt.show() # there is a machine precision order error which grows to factor*total_steps * 1e-19
    
    plt.title(f'|m_1|-1, N={N}')
    plt.plot(np.linalg.norm(Ynv,axis=1)-1)
    plt.plot(np.linalg.norm(Ynp,axis=1)-1)
    plt.plot(np.linalg.norm(Yg,axis=1)-1)
    plt.show()
```
